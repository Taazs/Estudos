```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
warnings.filterwarnings("ignore", category=UserWarning)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
```

```python
sinasc.columns
```

```python
sinasc['IDADEMAE'].describe()
```

```python
sns.boxplot(y = 'IDADEMAE', data = sinasc)
plt.title('Boxplot Idade da Mãe')
plt.ylabel('Idade da Mãe')
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(y = 'PESO', data = sinasc)
plt.title('Boxplot Peso do Bebê')
plt.ylabel('Peso do Bebê')
```

####  3) Faça um histograma da idade da mãe

```python
sinasc['IDADEMAE'].plot.hist(bins = 50)
plt.title('Histograma Idade Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')
```

```python
sns.displot(data = sinasc
           ,alpha = .5
           ,x = 'IDADEMAE'
           ,bins = 50
           ,element = 'step'
           ,kde = True
          )
```

#### 4) Faça um histograma do peso do bebê

```python
sinasc['PESO'].plot.hist(bins = 30)
plt.title('Histograma Peso do Bebê')
plt.xlabel('Peso')
plt.ylabel('Frequência')
```

```python
sns.displot(data = sinasc
           ,alpha = .25
           ,x = 'PESO'
           ,bins = 30
           ,element = 'step'
           ,kde = True
          )
```
